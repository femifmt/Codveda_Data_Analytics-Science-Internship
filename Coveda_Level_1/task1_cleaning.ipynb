```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Show all column
pd.set_option('display.max_columns', None)
```

```python
# Load the iris dataset
df_iris = pd.read_csv('iris.csv')
```

```python
# check the Shape of the dataframe 
print('shape :', df_iris)
# list of col
print("\nColumns:", df_iris.columns.tolist())
# Check first 5 rows of the dataset
print("\nFirst 5 rows:")
display(df_iris.head())
# Check datatypes
print("\nData types:\n", df_iris.dtypes)
```

```python
# Dataset Snapshot
def dataset_snapshot(df, name="Dataset"):
    # Add a divider line
    print(f" {'=' *50}")
    print(f" {name} Snapshot")
    print(f"{'=' *50}")
   
    # Number of rows
    print(f" Rows : {df.shape[0]}")
    
    # Number of columns
    print(f" Columns : {df.shape[1]}")
    
    # Check for Missing Values
    print(f" Missing : {df.isnull().sum().sum()}")
    
    # Check for Duplicate
    print(f" Duplicates : {df.duplicated().sum()}")
    # check memory usage
    print(f" Memory : {df.memory_usage(deep=True).sum() / 1024:.2f} KB")
    # Checks Data types
    print(f"\nData types:")
    print(df.dtypes)
    # Checks the first few rows
    print(f"\nFirst few rows:")
    display(df.head())
    
dataset_snapshot(df_iris, "Iris")
```

```python
# Show the actual duplicate rows
print("Duplicate rows in rae iris.csv:")
display(df_iris[df_iris.duplicated(keep='first')])
```

```python
# Create a copy of df_iris
df_iris_clean = df_iris.copy()
```

```python
# Removes deplicates
df_iris_clean = df_iris_clean.drop_duplicates()

# Standardize species column
# remove whitespaces and convert to lower case
df_iris_clean['species'] = df_iris_clean['species'].str.strip().str.lower()
```

```python
dataset_snapshot(df_iris_clean, "Iris")
```

```python
# Save clean iris Dataset as csv
df_iris_clean.to_csv('iris_clean.csv', index=False)
```
